```python
import glob
# Specify the list of samples to include
samples = [1, 2, 3, 5, 6, 8, 9, 10, 12, 14, 15, 17, 18, 19, 20]

# Define the file pattern for Prescription Drug Events (PDE)
opc_file_pattern = '/courses/ALY6980.202515/data/DE1_0_2008_to_2010_Outpatient_Claims_Sample_*.csv'

# Fetch all files matching the pattern
all_opc_files = sorted(glob.glob(opc_file_pattern))

# Filter files based on the desired sample numbers
selected_opc_files = [
    file for file in all_opc_files
    if any(f"Sample_{sample}" in file for sample in samples)
]

# Load and concatenate the selected files into one DataFrame
opc_data_frames = [pd.read_csv(file, low_memory=False) for file in selected_opc_files]
OPC_Final = pd.concat(opc_data_frames, ignore_index=True)

# Select only the desired columns
OPC_Final = OPC_Final[['DESYNPUF_ID', 'AT_PHYSN_NPI', 'OP_PHYSN_NPI', 'OT_PHYSN_NPI']]
```

Data Cleaning 

```python
# Convert SRVC_DT to datetime format to extract the year
PDE_Final['SRVC_DT'] = pd.to_datetime(PDE_Final['SRVC_DT'], format='%Y%m%d')

# Extract the year from SRVC_DT and create a new column 'YEAR'
PDE_Final['YEAR'] = PDE_Final['SRVC_DT'].dt.year

# Group by DESYNPUF_ID and YEAR, and sum the required columns
PDE_FinalDF = PDE_Final.groupby(['DESYNPUF_ID', 'YEAR'], as_index=False).agg({
    'QTY_DSPNSD_NUM': 'sum',
    'TOT_RX_CST_AMT': 'sum',
    'PTNT_PAY_AMT': 'sum'
})

PDE_FinalDF.head(10)
```

```python
PDE_FinalDF.to_csv('/home/kruayatidee.w/PDE_FinalDF.csv', index=False)
```

```python
# Calculate distinct counts for each column grouped by DESYNPUF_ID
OPC_Final['AT_PHYSN_NPI_CNT'] = OPC_Final.groupby('DESYNPUF_ID')['AT_PHYSN_NPI'].transform(lambda x: x.nunique())
OPC_Final['OP_PHYSN_NPI_CNT'] = OPC_Final.groupby('DESYNPUF_ID')['OP_PHYSN_NPI'].transform(lambda x: x.nunique())
OPC_Final['OT_PHYSN_NPI_CNT'] = OPC_Final.groupby('DESYNPUF_ID')['OT_PHYSN_NPI'].transform(lambda x: x.nunique())

# Display the updated dataframe
OPC_Final.head(10)
```

```python
OPC_FinalDF = OPC_Final.drop_duplicates(subset=['DESYNPUF_ID'])

OPC_FinalDF = OPC_FinalDF[['DESYNPUF_ID', 'AT_PHYSN_NPI_CNT', 'OP_PHYSN_NPI_CNT', 'OT_PHYSN_NPI_CNT']]

# Display the final dataset
OPC_FinalDF.head(10)
```

```python
import pandas as pd

# Convert CLM_FROM_DT from 'yyyymmdd' format to datetime
FINAL_M12['CLM_FROM_DT'] = pd.to_datetime(FINAL_M12['CLM_FROM_DT'], format='%Y%m%d')

# Sort by patient ID and claim date
FINAL_M12 = FINAL_M12.sort_values(by=['DESYNPUF_ID', 'CLM_FROM_DT'])

# Define a function to calculate prior payments within 6 months
def calculate_prior_payments(group):
    group = group.sort_values('CLM_FROM_DT')  # Ensure the group is sorted by date
    group['Previous_Payment_6M'] = group.apply(
        lambda row: group[
            (group['CLM_FROM_DT'] < row['CLM_FROM_DT']) &  # Prior dates
            (group['CLM_FROM_DT'] >= row['CLM_FROM_DT'] - pd.DateOffset(months=6))
        ]['CLM_PMT_AMT'].sum(),
        axis=1
    )
    return group

# Apply the calculation to each patient group
Final_M12 = FINAL_M12.groupby('DESYNPUF_ID').apply(calculate_prior_payments)

# Reset the index if necessary
Final_M12.reset_index(drop=True, inplace=True)

# Display the resulting dataframe
print(Final_M12[['DESYNPUF_ID', 'CLM_FROM_DT', 'CLM_PMT_AMT', 'Previous_Payment_6M']].head())
```

Grouped Bar Chart of Gender Proportion of each Chronic Disorder in each year

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Data Preparation
# Remove duplicates based on 'DESYNPUF_ID' and 'CLM_ADMSN_DT'
FINAL_HEARTDF_unique = final_heartdf.drop_duplicates(subset=['DESYNPUF_ID', 'CLM_ADMSN_DT'])

# Extract the year from 'CLM_ADMSN_DT' (assuming it's in 'yyyymmdd' format)
FINAL_HEARTDF_unique['ADMSN_YEAR'] = FINAL_HEARTDF_unique['CLM_ADMSN_DT'].astype(str).str[:4]

# Filter for relevant years and chronic diseases
years_of_interest = ['2008', '2009', '2010']
chronic_diseases = ["heart_failure", "copd", "ischemic_heart_disease", "stroke", "other"]

FINAL_HEARTDF_filtered = FINAL_HEARTDF_unique[
    (FINAL_HEARTDF_unique['ADMSN_YEAR'].isin(years_of_interest)) &
    (FINAL_HEARTDF_unique['FIRST_DGNS_CD'].isin(chronic_diseases))
]

# Define the color palette
palette = {
    "heart_failure": {"male": "skyblue", "female": "darkblue"},
    "copd": {"male": "salmon", "female": "red"},
    "ischemic_heart_disease": {"male": "lightgreen", "female": "darkgreen"},
    "stroke": {"male": "orange", "female": "darkorange"},
    "other": {"male": "plum", "female": "purple"}
}

# Step 2: Grouped Bar Chart
fig, axes = plt.subplots(3, 2, figsize=(18, 12), sharey=True)
axes = axes.flatten()

# Loop through diseases to create grouped bar charts
for i, disease in enumerate(chronic_diseases):
    if i >= 6:  # Stop after filling a 3x2 grid
        break

    # Filter data for each disease
    disease_data = FINAL_HEARTDF_filtered[FINAL_HEARTDF_filtered['FIRST_DGNS_CD'] == disease]
    
    # Calculate proportions of gender for each year
    proportion_data = (disease_data.groupby(['ADMSN_YEAR', 'BENE_SEX_IDENT_CD'])
                                    .size()
                                    .unstack(fill_value=0))
    proportion_data = proportion_data.div(proportion_data.sum(axis=1), axis=0)  # Normalize to proportions
    
    # Create a grouped bar chart
    ax = axes[i]
    proportion_data.plot(kind='bar', ax=ax, width=0.6,  # Reduced bar width
                         color=[palette[disease]["female"], palette[disease]["male"]])
    ax.set_title(disease.replace('_', ' ').title())
    ax.set_xlabel("Year")
    ax.set_ylabel("Proportion")
    ax.legend(["Female", "Male"], title="Sex")
    
    # Add labels above bars
    for p in ax.patches:
        if p.get_height() > 0:  # Only label non-zero bars
            ax.annotate(f"{p.get_height():.2f}",
                        (p.get_x() + p.get_width() / 2, p.get_height()),
                        ha='center', va='bottom', fontsize=8)

# Hide unused subplots if there are fewer than 6 diseases
axes[-1].axis('off')

# Adjust layout
plt.tight_layout(pad=1.5)
plt.subplots_adjust(top=0.95)
fig.suptitle("Gender Proportion by Year and Chronic Disorder (2008-2010)", fontsize=18, y=0.98)
plt.show()
```

```python
!pip install dash

import pandas as pd
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots 

# Assuming FINAL_HEARTDF is already loaded in the environment

# Remove duplicates based on 'DESYNPUF_ID' and 'CLM_ADMSN_DT'
df_unique = final_heartdf.drop_duplicates(subset=['DESYNPUF_ID', 'CLM_ADMSN_DT'])

# Extract the year from 'CLM_ADMSN_DT' (assuming it's a string and can be sliced)
df_unique['ADMSN_YEAR'] = df_unique['CLM_ADMSN_DT'].astype(str).str[:4]

# Define the list of years and chronic disorders
years_of_interest = ['2008', '2009', '2010']
chronic_diseases = ["heart_failure", "copd", "ischemic_heart_disease", "stroke", "other"]

# Mapping for sex labels (assuming 1=Male, 2=Female)
sex_labels = {1: "Male", 2: "Female"}
df_unique['BENE_SEX_IDENT_CD'] = df_unique['BENE_SEX_IDENT_CD'].map(sex_labels)

# Define the color palette for non-gender-specific charts
palette = {
    "heart_failure": "skyblue",
    "copd": "salmon",
    "ischemic_heart_disease": "lightgreen",
    "stroke": "orange",
    "other": "plum"
}

# Define the gender-specific palette for the gender bar chart
gender_palette = {
    "heart_failure": {"male": "skyblue", "female": "darkblue"},
    "copd": {"male": "salmon", "female": "red"},
    "ischemic_heart_disease": {"male": "lightgreen", "female": "darkgreen"},
    "stroke": {"male": "orange", "female": "darkorange"},
    "other": {"male": "plum", "female": "purple"}
}

# Initialize the Dash app
app = Dash(__name__)

app.layout = html.Div([
    html.H1("Heart Disease Dashboard"),
    
    # Dropdown for selecting year
    html.Label("Select Year:"),
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': year, 'value': year} for year in years_of_interest] + [{'label': 'Select All', 'value': 'select_all'}],
        value='2008'
    ),
    
    # Dropdown for selecting chronic disorder, with "Select All" option
    html.Label("Select Chronic Disorder:"),
    dcc.Dropdown(
        id='disorder-dropdown',
        options=[{'label': disease.replace("_", " ").title(), 'value': disease} for disease in chronic_diseases] + [{"label": "Select All", "value": "select_all"}],
        value='select_all'
    ),
    
    # Pie chart for disease proportions
    dcc.Graph(id='pie-chart'),
    
    # Box plot for total admission days by year and disease
    dcc.Graph(id='box-plot'),
    
    # Bar chart for gender proportion by year and disorder
    dcc.Graph(id='gender-bar-chart'),

    # Line chart for age distribution by year and chronic disorder
    dcc.Graph(id='age-distribution-line-chart')
])

@app.callback(
    [Output('pie-chart', 'figure'),
     Output('box-plot', 'figure'),
     Output('gender-bar-chart', 'figure'),
     Output('age-distribution-line-chart', 'figure')],
    [Input('year-dropdown', 'value'),
     Input('disorder-dropdown', 'value')]
)
def update_charts(selected_year, selected_disorder):
    # Handle filtering for year
    if selected_year == 'select_all':
        data_year = df_unique[df_unique['ADMSN_YEAR'].isin(years_of_interest)]
    else:
        data_year = df_unique[df_unique['ADMSN_YEAR'] == selected_year]

    # Handle filtering for disorder
    if selected_disorder == 'select_all':
        data_disorder = data_year
    else:
        data_disorder = data_year[data_year['FIRST_DGNS_CD'] == selected_disorder]

    # Pie Chart: Proportion of First Diagnosed Diseases
    if selected_year == 'select_all':
        # Generate one pie chart for each year
        fig = make_subplots(rows=1, cols=3, specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]],
                            subplot_titles=years_of_interest)
        for i, year in enumerate(years_of_interest):
            yearly_data = data_year[data_year['ADMSN_YEAR'] == year]
            disease_counts = yearly_data['FIRST_DGNS_CD'].value_counts().reindex(chronic_diseases, fill_value=0)
            fig.add_trace(go.Pie(
                labels=chronic_diseases,
                values=disease_counts,
                marker=dict(colors=[palette[disease] for disease in chronic_diseases])
            ), 1, i+1)
        fig.update_layout(
            title="Proportion of First Diagnosed Diseases by Year",
            showlegend=True
        )
        pie_chart = fig
    else:
        # Single pie chart for the selected year
        disease_counts = data_year['FIRST_DGNS_CD'].value_counts().reindex(chronic_diseases, fill_value=0)
        pie_chart = go.Figure(
            data=[go.Pie(
                labels=chronic_diseases, 
                values=disease_counts,
                marker=dict(colors=[palette[disease] for disease in chronic_diseases])
            )]
        )
        pie_chart.update_layout(title_text=f"Proportion of First Diagnosed Diseases in {selected_year}")

    # Box Plot: Total Admission Days
    box_plot = px.box(
        data_disorder, 
        x='FIRST_DGNS_CD', 
        y='TOTAL_ADMSN_DY',
        color='FIRST_DGNS_CD',
        color_discrete_map=palette,
        title=f"Total Admission Days ({'All Years' if selected_year == 'select_all' else selected_year}) ({'All Disorders' if selected_disorder == 'select_all' else selected_disorder.replace('_', ' ').title()})"
    )
    box_plot.update_layout(xaxis_title="Chronic Disorder", yaxis_title="Total Admission Days")

    # Gender Proportion Bar Chart
    if selected_disorder == "select_all":
        gender_prop_data = (data_year.groupby(['FIRST_DGNS_CD', 'BENE_SEX_IDENT_CD'])
                                         .size()
                                         .unstack(fill_value=0))
        gender_prop_data = gender_prop_data.div(gender_prop_data.sum(axis=1), axis=0).reset_index()
        gender_prop_data = gender_prop_data.melt(id_vars='FIRST_DGNS_CD', value_name='Proportion', var_name='Sex')

        gender_bar_chart = px.bar(
            gender_prop_data,
            x='FIRST_DGNS_CD',
            y='Proportion',
            color='Sex',
            barmode='group',
            title=f"Proportion of Sex by Disorder ({'All Years' if selected_year == 'select_all' else selected_year})",
            color_discrete_map={
                "Female": "deepskyblue",
                "Male": "darkblue"
            }
        )
        gender_bar_chart.update_traces(marker_line_width=0)
        gender_bar_chart.update_layout(xaxis_title="Chronic Disorder", yaxis_title="Proportion", bargap=0.15)
    else:
        gender_prop_data = (data_disorder.groupby(['ADMSN_YEAR', 'BENE_SEX_IDENT_CD'])
                                         .size()
                                         .unstack(fill_value=0))
        gender_prop_data = gender_prop_data.div(gender_prop_data.sum(axis=1), axis=0).reset_index()
        gender_prop_data = gender_prop_data.melt(id_vars='ADMSN_YEAR', value_name='Proportion', var_name='Sex')

        gender_bar_chart = px.bar(
            gender_prop_data,
            x='ADMSN_YEAR',
            y='Proportion',
            color='Sex',
            barmode='group',
            title=f"Proportion of Sex for {selected_disorder.replace('_', ' ').title()} ({'All Years' if selected_year == 'select_all' else selected_year})",
            color_discrete_map={
                "Female": gender_palette[selected_disorder]["female"],
                "Male": gender_palette[selected_disorder]["male"]
            }
        )
        gender_bar_chart.update_traces(marker_line_width=0)
        gender_bar_chart.update_layout(xaxis_title="Year", yaxis_title="Proportion", bargap=0.15)

    # Age Distribution Line Chart
    age_distribution_chart = go.Figure()
    if selected_disorder == "select_all":
        for disorder in chronic_diseases:
            age_counts = data_year[data_year['FIRST_DGNS_CD'] == disorder]['AGE_YEARS'].value_counts().sort_index()
            age_distribution_chart.add_trace(go.Scatter(
                x=age_counts.index, 
                y=age_counts.values, 
                mode='lines+markers', 
                name=f"{disorder.replace('_', ' ').title()}",
                line=dict(color=palette[disorder])
            ))
    else:
        for year in years_of_interest:
            if selected_year != 'select_all' and year != selected_year:
                continue
            age_counts = data_disorder[data_disorder['ADMSN_YEAR'] == year]['AGE_YEARS'].value_counts().sort_index()
            age_distribution_chart.add_trace(go.Scatter(
                x=age_counts.index, 
                y=age_counts.values, 
                mode='lines+markers', 
                name=f"{year}",
                line=dict(color=palette[selected_disorder])
            ))
    age_distribution_chart.update_layout(title=f"Age Distribution ({'All Disorders' if selected_disorder == 'select_all' else selected_disorder.replace('_', ' ').title()})", xaxis_title="Age", yaxis_title="Frequency")

    return pie_chart, box_plot, gender_bar_chart, age_distribution_chart


# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
# Compute approximated AIC
n = len(y_test)  # Number of observations
rss = np.sum((y_test - y_pred_rfot) ** 2)  # Residual Sum of Squares
k = model_rfot.n_estimators  # Use the number of trees as a proxy for the number of parameters
log_likelihood = -0.5 * n * np.log(rss / n)  # Log-likelihood approximation
aic = 2 * k - 2 * log_likelihood  # AIC formula
print("Approximated AIC:", aic)
```
